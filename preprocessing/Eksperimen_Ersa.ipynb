# **5. Data Preprocessing**

Pada tahap ini, data preprocessing adalah langkah penting untuk memastikan kualitas data sebelum digunakan dalam model machine learning.

```python
# 5. Distribusi kelas target (quality)
print("\n--- Distribusi Kelas Target ---")
print(df['quality'].value_counts(normalize=True))
```

```python
# 1. Menghapus data duplikat
df_clean = df.drop_duplicates().copy()
print(f"Banyak data setelah menghapus duplikat: {df_clean.shape[0]} (sebelumnya: {df.shape[0]})")
```

```python
# 2. Menangani missing values dengan median
impute_cols = ['citric_acid', 'pH']
for col in impute_cols:
    median_val = df_clean[col].median()
    df_clean[col] = df_clean[col].fillna(median_val)
    print(f"Mengisi nilai kosong pada '{col}' dengan nilai median: {median_val:.4f}")

print("Jumlah nilai kosong setelah pembersihan:")
print(df_clean.isnull().sum())
```

```python
# 3. Standarisasi Fitur Numerik
features = df_clean.drop(columns=['quality'])
target = df_clean['quality']

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Konversi kembali ke DataFrame
df_preprocessed = pd.DataFrame(scaled_features, columns=features.columns)
df_preprocessed['quality'] = target.reset_index(drop=True)

print("Dimensi data hasil preprocessing:", df_preprocessed.shape)
df_preprocessed.head()
```

```python
# 4. Menyimpan dataset hasil preprocessing
output_dir = "./namadataset_preprocessing"
os.makedirs(output_dir, exist_ok=True)

output_path = os.path.join(output_dir, "wine_preprocessed.csv")
df_preprocessed.to_csv(output_path, index=False)

# Simpan juga langsung ke level preprocessing untuk kemudahan akses model
df_preprocessed.to_csv("./wine_preprocessed.csv", index=False)

print(f"Dataset preprocessing berhasil disimpan ke {output_path}!")
```
